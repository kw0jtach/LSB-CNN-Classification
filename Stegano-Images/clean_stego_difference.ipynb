```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_normal = cv2.imread('stegoimagesdataset/train/train/clean/00001.png')
img_stego_eth_0 = cv2.imread('stegoimagesdataset/train/train/stego/image_00001_eth_0.png')
img_stego_eth_1 = cv2.imread('stegoimagesdataset/train/train/stego/image_00001_eth_1.png')
img_stego_ps = cv2.imread('stegoimagesdataset/train/train/stego/image_00001_ps_0.png')

lsb_normal = img_normal & 1
lsb_normal = lsb_normal * 255
lsb_stego_eth_0 = img_stego_eth_0 & 1
lsb_stego_eth_0 = lsb_stego_eth_0 * 255
lsb_stego_eth_1 = img_stego_eth_1 & 1
lsb_stego_eth_1 = lsb_stego_eth_1 * 255
lsb_stego_ps = img_stego_ps & 1
lsb_stego_ps = lsb_stego_ps * 255

plt.figure(figsize=(10, 5))
plt.subplot(2, 2, 1)
plt.imshow(lsb_normal, cmap='gray')
plt.title('LSB of Normal Image')
plt.axis('off')
plt.subplot(2, 2, 2)
plt.imshow(lsb_stego_eth_0, cmap='gray')
plt.title('LSB of Stego Image ETH 0')
plt.axis('off')
plt.subplot(2, 2, 3)
plt.imshow(lsb_stego_eth_1, cmap='gray')
plt.title('LSB of Stego Image ETH 1')
plt.axis('off')
plt.subplot(2, 2, 4)
plt.imshow(lsb_stego_ps, cmap='gray')
plt.title('LSB of Stego Image PS')
plt.axis('off')
plt.tight_layout()
plt.show()

# Laplcian
lap_normal = cv2.Laplacian(img_normal, cv2.CV_64F)
lap_normal = np.uint8(np.absolute(lap_normal))
lap_stego_eth_0 = cv2.Laplacian(img_stego_eth_0, cv2.CV_64F)
lap_stego_eth_0 = np.uint8(np.absolute(lap_stego_eth_0))
lap_stego_eth_1 = cv2.Laplacian(img_stego_eth_1, cv2.CV_64F)
lap_stego_eth_1 = np.uint8(np.absolute(lap_stego_eth_1))
lap_stego_ps = cv2.Laplacian(img_stego_ps, cv2.CV_64F)
lap_stego_ps = np.uint8(np.absolute(lap_stego_ps))
plt.figure(figsize=(10, 5))
plt.subplot(2, 2, 1)
plt.imshow(lap_normal, cmap='gray')
plt.title('Laplacian of Normal Image')
plt.axis('off')
plt.subplot(2, 2, 2)
plt.imshow(lap_stego_eth_0, cmap='gray')
plt.title('Laplacian of Stego Image ETH 0')
plt.axis('off')
plt.subplot(2, 2, 3)
plt.imshow(lap_stego_eth_1, cmap='gray')
plt.title('Laplacian of Stego Image ETH 1')
plt.axis('off')
plt.subplot(2, 2, 4)
plt.imshow(lap_stego_ps, cmap='gray')
plt.title('Laplacian of Stego Image PS')
plt.axis('off')
plt.tight_layout()
plt.show()

# histograms of laplacian
plt.figure(figsize=(10, 5))
plt.subplot(2, 2, 1)
plt.hist(lap_normal.ravel(), 256, [0, 256])
plt.title('Histogram of Laplacian of Normal Image')
plt.subplot(2, 2, 2)
plt.hist(lap_stego_eth_0.ravel(), 256, [0, 256])
plt.title('Histogram of Laplacian of Stego Image ETH 0')
plt.subplot(2, 2, 3)
plt.hist(lap_stego_eth_1.ravel(), 256, [0, 256])
plt.title('Histogram of Laplacian of Stego Image ETH 1')
plt.subplot(2, 2, 4)
plt.hist(lap_stego_ps.ravel(), 256, [0, 256])
plt.title('Histogram of Laplacian of Stego Image PS')
plt.tight_layout()
plt.show()
```
